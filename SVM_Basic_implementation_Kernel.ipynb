### SVM Kernels Indepth Intuition And Practical Explanation

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5.0, 5.0, 100)
y = np.sqrt(10**2 - x**2)
y=np.hstack([y,-y])
x=np.hstack([x,-x])
```

x = np.linspace(-5.0, 5.0, 100)
Meaning: Creates an array of 100 evenly spaced values between -5.0 and 5.0.
Purpose: These values represent the x-coordinates for points on a circle.
Mathematical intuition: This is a set of x-values that will be used to compute corresponding y-values on a circle.


y = np.sqrt(10**2 - x**2)
Meaning: For each x-value, computes the positive y-value that satisfies the equation of a circle: ( x^2 + y^2 = r^2 ), where ( r = 10 ).
Purpose: Generates the upper half of a circle with radius 10 centered at the origin.
Mathematical intuition: For each x, ( y = \sqrt{r^2 - x^2} ) gives the upper semicircle.



y = np.hstack([y, -y])
Meaning: Stacks the positive y-values (upper semicircle) and their negatives (lower semicircle) into a single array.
Purpose: Creates y-coordinates for the full circle (both upper and lower halves).
Mathematical intuition: For each x, both ( +y ) and ( -y ) are valid solutions to the circle equation.


x = np.hstack([x, -x])
Meaning: Stacks the original x-values and their negatives into a single array.
Purpose: Ensures that for every y (positive and negative), there is a corresponding x, effectively mirroring the points to complete the circle.
Mathematical intuition: This step is used to pair the y-values with both the original and mirrored x-values, ensuring a full circle is represented.



This code generates the (x, y) coordinates for points on a circle of radius 10 centered at the origin. The approach first computes the upper semicircle, then mirrors the points to get the lower semicircle, and finally mirrors the x-values to ensure all points on the circle are covered. This is a common way to generate synthetic data for visualizing non-linear boundaries, such as those used in SVM kernel demonstrations.

```python
## 1. Generate Points for Two Circles

x1 = np.linspace(-5.0, 5.0, 100)
y1 = np.sqrt(5**2 - x1**2)
y1=np.hstack([y1,-y1])
x1=np.hstack([x1,-x1])
```

Meaning:

x1 = np.linspace(-5.0, 5.0, 100): Generates 100 evenly spaced x-values between -5 and 5.
y1 = np.sqrt(5**2 - x1**2): For each x1, computes the positive y-value on a circle of radius 5 centered at (0,0) using ( x^2 + y^2 = r^2 ).
y1 = np.hstack([y1, -y1]): Stacks the positive and negative y-values to get both the upper and lower halves of the circle.
x1 = np.hstack([x1, -x1]): Duplicates x1 values to match the length of y1.
Purpose:
To create the coordinates for a small circle (radius 5) for later use as one class in a classification problem.

Mathematical Intuition:
This generates points on a circle, which are not linearly separable from another concentric circle in 2D.

```python
##2. Generate Points for a Larger Circle

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5.0, 5.0, 100)
y = np.sqrt(10**2 - x**2)
y=np.hstack([y,-y])
x=np.hstack([x,-x])

x1 = np.linspace(-5.0, 5.0, 100)
y1 = np.sqrt(5**2 - x1**2)
y1=np.hstack([y1,-y1])
x1=np.hstack([x1,-x1])

plt.scatter(y,x)
plt.scatter(y1,x1)
```

Meaning:

Imports NumPy and Matplotlib for numerical operations and plotting.
Generates points for a large circle (radius 10) in the same way as above.
Also regenerates the small circle.
plt.scatter(y, x) and plt.scatter(y1, x1): Plots both circles.
Purpose:
To visualize two concentric circles, which will be used as two classes for SVM classification.

Mathematical Intuition:
These two circles are not linearly separable in 2D, making them a classic example for kernel SVMs.

```python
## 3. Stack Data for DataFrame Creation

np.vstack([y,x]).T
```

Meaning:

Stacks y and x arrays vertically and transposes them, so each row is a point [y, x].
Purpose:
To prepare the data for DataFrame creation.

```python
## 4. Create DataFrames for Each Class and Combine


import pandas as pd
df1 =pd.DataFrame(np.vstack([y,x]).T,columns=['X1','X2'])
df1['Y']=0
df2 =pd.DataFrame(np.vstack([y1,x1]).T,columns=['X1','X2'])
df2['Y']=1
df = pd.concat([df1, df2], ignore_index=True)
df.head(5)
```

Meaning:

Imports pandas for data manipulation.
Creates DataFrames for each circle, labeling the large circle as class 0 and the small circle as class 1.
Appends both DataFrames to create a single dataset.
Displays the first 5 rows.
Purpose:
To create a labeled dataset for binary classification.

### Polynomial Kernel

```python
# We need to find components for the Polynomical Kernel
#X1,X2,X1_square,X2_square,X1*X2
## 8. Add Polynomial Features

df['X1_Square']= df['X1']**2
df['X2_Square']= df['X2']**2
df['X1*X2'] = (df['X1'] *df['X2'])
df.head()
```

Meaning:

Adds new columns: X1 squared, X2 squared, and X1*X2.
Purpose:
To explicitly map the data to a higher-dimensional space, mimicking what a polynomial kernel does implicitly.

Mathematical Intuition:
Polynomial kernels allow SVMs to find non-linear boundaries by mapping data to higher dimensions.

```python
### Independent and Dependent features
## 9. Select New Feature Set

X = df[['X1','X2','X1_Square','X2_Square','X1*X2']]
y = df['Y']
```

Purpose:
To use the new polynomial features for classification.

```python
## 10. Split New Feature Set
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size = 0.25,
                                                    random_state = 0)
```

```python
## 14. Train and Evaluate Linear SVM

from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
classifier = SVC(kernel="linear")
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
accuracy_score(y_test, y_pred)
```

Meaning:

Imports SVC and accuracy metric.

Predicts test labels and computes accuracy.
Purpose:
To see if a linear SVM can now separate the classes in the higher-dimensional space.

Mathematical Intuition:
Linear SVMs find a hyperplane in the feature space. With polynomial features, the data may become linearly separable.

```python
## 15. Train and Evaluate Polynomial Kernel SVM

from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
classifier = SVC(kernel="poly")
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
accuracy_score(y_test, y_pred)
```

Purpose:
To use the polynomial kernel, which implicitly maps data to a higher-dimensional space.

Mathematical Intuition:
The polynomial kernel computes ( (x \cdot x' + c)^d ), allowing the SVM to find non-linear boundaries.

```python
## 16. Train and Evaluate RBF Kernel SVM

from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
classifier = SVC(kernel="rbf")
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
accuracy_score(y_test, y_pred)
```

Purpose:
To use the RBF (Gaussian) kernel, which maps data to infinite-dimensional space.

Mathematical Intuition:
The RBF kernel computes ( \exp(-\gamma |x - x'|^2) ), allowing the SVM to create highly flexible, non-linear boundaries.

```python
## 17. Train and Evaluate Sigmoid Kernel SVM

from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
classifier = SVC(kernel="sigmoid")
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
accuracy_score(y_test, y_pred)
```

Purpose:
To use the sigmoid kernel, which mimics the behavior of a neural network activation function.

Mathematical Intuition:
The sigmoid kernel computes ( \tanh(\alpha x \cdot x' + c) ), introducing non-linearity similar to neural networks.
