logistic=LogisticRegression(): Initializes a Logistic Regression model. By default, it uses an L2 penalty (regularization), 'lbfgs' solver, and 'ovr' (One-vs-Rest) strategy for multiclass (though not strictly necessary for binary, it works).

logistic.fit(X_train,y_train): This is the core training step. The model learns the relationship between the features (X_train) and the target variable (y_train).

Mathematical Intuition: Logistic Regression estimates the probability that a given input belongs to a particular class. It does this by passing the linear combination of inputs (features) and their weights through a sigmoid (logistic) function. The sigmoid function squashes the output of the linear model to a value between 0 and 1, which can be interpreted as a probability.

The linear model is: z = b0 + b1*x1 + b2*x2 + ... + bn*xn (where b are coefficients/weights and x are features).

The sigmoid function is: P(y=1|X) = 1 / (1 + e^(-z))

During fit(), the model uses an optimization algorithm (like 'lbfgs' by default) to find the optimal coefficients (b0, b1, ... bn) that minimize the log-loss (also known as cross-entropy loss) between the predicted probabilities and the actual class labels in the training data.

y_pred=logistic.predict(X_test): After training, this line uses the trained model to make predictions on the unseen X_test data.

Mathematical Intuition: For each data point in X_test, the model calculates the probability P(y=1|X) using the learned coefficients. If this probability is above a certain threshold (default is 0.5), the model predicts class 1; otherwise, it predicts class 0.

Mathematical Intuition: This report includes:

Precision: True Positives / (True Positives + False Positives). It answers: "Of all instances predicted as positive, how many were actually positive?"

Recall (Sensitivity): True Positives / (True Positives + False Negatives). It answers: "Of all actual positive instances, how many did the model correctly identify?"

F1-score: The harmonic mean of precision and recall, providing a single metric that balances both.

Support: The number of actual occurrences of each class in the y_test set.

print(confusion_matrix(y_pred,y_test)): Computes a confusion matrix.

Mathematical Intuition: A confusion matrix is a table that is often used to describe the performance of a classification model on a set of test data for which the true values are known. It allows visualization of the performance of an algorithm. The rows correspond to predicted classes and columns correspond to actual classes (or vice versa).

[[True Negatives, False Positives],[False Negatives, True Positives]

## Hyperparameter Tuning And Cross Validation

## Grid SearchCV

model=LogisticRegression(): Initializes a new LogisticRegression estimator, which will be used as the base model for GridSearchCV.

penalty=['l1', 'l2', 'elasticnet']: Defines a list of regularization penalties to try.

Mathematical Intuition (Regularization): Regularization adds a penalty to the loss function to prevent overfitting by discouraging overly complex models (i.e., models with very large coefficients).

L1 Regularization (Lasso): Adds a penalty proportional to the absolute value of the coefficients (sum(|coefficients|)). It can lead to sparse models where some coefficients become exactly zero, effectively performing feature selection.

L2 Regularization (Ridge): Adds a penalty proportional to the square of the magnitude of the coefficients (sum(coefficients^2)). It shrinks coefficients towards zero but doesn't typically make them exactly zero.

Elastic-Net Regularization: A hybrid of L1 and L2 regularization, combining their benefits.

c_values=[100,10,1.0,0.1,0.01]: Defines a list of C values to try.

Mathematical Intuition (C parameter): C is the inverse of regularization strength. Smaller values of C specify stronger regularization (more penalty), while larger values of C specify weaker regularization (less penalty). It helps control the trade-off between fitting the training data well and keeping the model simple.

solver=['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']: Defines a list of optimization algorithms (solvers) to try.

Mathematical Intuition (Solvers): These algorithms are used to minimize the cost function (log-loss) during the training of the Logistic Regression model. Different solvers are more efficient for different types of problems and penalty types.

liblinear: Good for small datasets, supports L1 and L2 penalties.

newton-cg, lbfgs, sag: Good for larger datasets, support L2 and 'none' penalties.

saga: Supports all penalties (L1, L2, Elastic-Net, none) and is good for large datasets.

```python
model=LogisticRegression()
penalty=['l1', 'l2', 'elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver=['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver)
```

params=dict(penalty=penalty,C=c_values,solver=solver): Creates a dictionary params where keys are the hyperparameter names ('penalty', 'C', 'solver') and values are the lists of values to be tested for each hyperparameter. This dictionary defines the "grid" for GridSearchCV.

```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
```

from sklearn.model_selection import StratifiedKFold: Imports StratifiedKFold for cross-validation.

cv=StratifiedKFold(): Initializes StratifiedKFold cross-validation.

Purpose: StratifiedKFold is a variation of K-Fold cross-validation that ensures each fold has approximately the same percentage of samples for each target class as the complete set. This is particularly important for imbalanced datasets to ensure that each fold is representative. By default, n_splits is 5.

Mathematical Intuition (Cross-Validation): Instead of a single train-test split, cross-validation divides the data into k folds. The model is trained k times, each time using k-1 folds for training and 1 fold for testing. The performance scores from each fold are then averaged, providing a more robust estimate of the model's performance and reducing the variance associated with a single train-test split.

```python
## GridSearchCV
from sklearn.model_selection import GridSearchCV

grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

from sklearn.model_selection import GridSearchCV: Imports the GridSearchCV class.

grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1): Initializes the GridSearchCV object.

estimator=model: The LogisticRegression model initialized earlier.

param_grid=params: The dictionary defining the hyperparameter combinations to search.

scoring='accuracy': The metric used to evaluate the performance of each parameter combination during cross-validation. The goal is to maximize accuracy.

cv=cv: The cross-validation strategy to use (StratifiedKFold in this case).

n_jobs=-1: Uses all available CPU cores for parallel processing, speeding up the search.

Purpose: GridSearchCV will try every possible combination of penalty, C, and solver defined in params, train a model with each combination using StratifiedKFold cross-validation, and evaluate its performance based on 'accuracy'.

```python
grid
```

```python
grid.fit(X_train,y_train)
```

grid.fit(X_train,y_train): Executes the grid search. For each parameter combination, it performs cross-validation on X_train and y_train to find the best performing set of hyperparameters.

Mathematical Intuition: This process involves repeatedly training and evaluating the Logistic Regression model with different hyperparameter settings. For each setting, the model calculates probabilities via the sigmoid function and optimizes its coefficients to minimize log-loss. The scoring parameter (accuracy) guides the selection of the "best" set of hyperparameters by maximizing the average accuracy across all cross-validation folds.

```python
grid.best_params_
```

grid.best_params_: After fit(), this attribute returns a dictionary containing the hyperparameters that yielded the best score during the grid search.

```python
grid.best_score_
```

grid.best_score_: Returns the mean cross-validated score of the best estimator found during the grid search.

```python
y_pred=grid.predict(X_test)
```

y_pred=grid.predict(X_test): Makes predictions on the X_test data using the best LogisticRegression model found by GridSearchCV. This model is automatically retrained on the entire X_train dataset with the best_params_.

```python
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```

These lines re-evaluate the model's performance on the test set using the best hyperparameters found by GridSearchCV, providing the accuracy, classification report, and confusion matrix.

Mathematical Intuition (One-vs-Rest - OVR): For multiclass classification, Logistic Regression typically uses one of two strategies:

One-vs-Rest (OvR) / One-vs-All (OvA): This strategy trains a separate binary logistic regression model for each class. For a dataset with N classes, it trains N models. Each model distinguishes one class from all the others. For prediction, it runs all N models on the input, and the class with the highest predicted probability (or score) is chosen as the final prediction.

Multinomial (Softmax Regression): This strategy directly models the probability distribution of the classes, extending the sigmoid function to a softmax function for multiple classes. It's a single model that learns to differentiate between all classes simultaneously.

By explicitly setting multi_class='ovr', the model is configured to use the One-vs-Rest approach.

logistic.fit(X_train,y_train): Trains the multiclass logistic regression model on the training data.

Mathematical Intuition: If multi_class='ovr' is used, during fitting, it essentially trains 3 independent binary logistic regression models (since n_classes=3): one for class 0 vs. (1 and 2), one for class 1 vs. (0 and 2), and one for class 2 vs. (0 and 1). Each binary model minimizes its own log-loss.

from sklearn.model_selection import GridSearchCV

Purpose: This line imports the GridSearchCV class from scikit-learn's model_selection module. GridSearchCV is a powerful tool for hyperparameter tuning.

Mathematical Intuition: In machine learning, models often have hyperparameters (e.g., the regularization strength 'C' in Logistic Regression) that are not learned from the data but are set prior to training. The performance of a model can be highly sensitive to these hyperparameters. GridSearchCV automates the process of exhaustively searching for the best combination of hyperparameters from a specified set, by training and evaluating the model for each combination.

from sklearn.model_selection import StratifiedKFold

Purpose: This line imports the StratifiedKFold class, also from scikit-learn's model_selection module. It is a cross-validation strategy.

Mathematical Intuition: Cross-validation is a resampling procedure used to evaluate machine learning models on a limited data sample. K-Fold Cross-Validation divides the dataset into 'k' folds (subsets). The model is trained on k-1 folds and validated on the remaining fold. This process is repeated 'k' times, with each fold serving as the validation set exactly once. StratifiedKFold is particularly useful for classification tasks where the target variable classes are imbalanced. It ensures that each fold has approximately the same percentage of samples for each target class as the complete set. This helps prevent bias in the performance evaluation, as a random split might result in some folds having very few or no samples of a minority class.

cv=StratifiedKFold()

Purpose: This line creates an instance of the StratifiedKFold cross-validator. By default, it uses 5 splits (k=5).

Mathematical Intuition: This step initializes the strategy for dividing the data during the grid search. It prepares the cross-validation object that GridSearchCV will use internally to create training and validation sets for each combination of hyperparameters. The StratifiedKFold object ensures that each fold maintains the original proportion of target classes, which is crucial for robust model evaluation, especially with imbalanced datasets.

grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv)

Purpose: This is the core line for setting up the grid search.

estimator=model: Specifies the machine learning model to be tuned (in this case, the LogisticRegression model defined earlier).

param_grid=params: Provides the dictionary (params) containing the hyperparameter values to explore.

scoring='accuracy': Defines the evaluation metric to optimize during the grid search. The model that achieves the highest accuracy on the cross-validation folds will be chosen as the best model.

cv=cv: Sets the cross-validation strategy to the StratifiedKFold instance created in the previous line.

Mathematical Intuition: This line configures the GridSearchCV object. It tells GridSearchCV to systematically try every possible combination of hyperparameters listed in params. For each combination, it will perform k-fold cross-validation (as defined by cv). In each fold, the model is trained on k-1 parts of the data and evaluated on the remaining part using 'accuracy' as the metric. The GridSearchCV then aggregates these scores across all folds for each parameter combination. The combination of hyperparameters that yields the highest average 'accuracy' across all folds is considered the best and is chosen as the optimal set of hyperparameters for the model. This systematic search helps in finding the hyperparameters that lead to the best generalization performance of the model on unseen data.

```python
grid.fit(X_train,y_train)
```

Purpose: This line executes the grid search process. It fits the GridSearchCV object to the training data (X_train, y_train).

Mathematical Intuition: When fit is called, GridSearchCV performs the following iterative process:

Iterate through param_grid: For each combination of hyperparameters specified in params.

Perform Cross-Validation: For each hyperparameter combination, it performs the StratifiedKFold cross-validation:

Divides X_train and y_train into k (default 5) stratified folds.

In each of the k iterations:

One fold is held out as the validation set.

The remaining k-1 folds are used as the training set.

A new LogisticRegression model is initialized with the current hyperparameter combination.

This model is trained on the current training set.

The trained model's performance is evaluated on the current validation set using the 'accuracy' scoring metric.

Calculate Average Score: The accuracy scores from all k folds for that specific hyperparameter combination are averaged.

Identify Best Parameters: After evaluating all hyperparameter combinations, GridSearchCV identifies the combination that resulted in the highest average accuracy across its respective cross-validation folds.

Retrain Best Model: Finally, GridSearchCV automatically retrains a new LogisticRegression model on the entire X_train and y_train dataset using the best hyperparameters found. This final model is then stored in grid.best_estimator_. This step is crucial because the previous steps were for hyperparameter selection and model evaluation using cross-validation, but the final model used for prediction should be trained on all available training data to maximize its learning.

```python
y_pred=grid.predict(X_test)
```

Purpose: This line uses the best model (which was automatically retrained on the full training data with the optimal hyperparameters by GridSearchCV and stored as grid.best_estimator_) to make predictions on the X_test dataset.

Mathematical Intuition: The predict method of the fitted GridSearchCV object internally calls the predict method of its best_estimator_. For each data point x_i in X_test, the Logistic Regression model calculates the log-odds of the positive class using the learned coefficients (weights) and intercept: log_odds = b_0 + b_1*x_1 + ... + b_n*x_n. This log_odds value is then transformed into a probability using the sigmoid function: P(y=1|x) = 1 / (1 + e^(-log_odds)). Finally, a classification decision is made: if P(y=1|x) is greater than or equal to a default threshold (usually 0.5), the prediction is class 1; otherwise, it's class 0. These predictions form the y_pred array.

dummy_fpr, dummy_tpr, _ = roc_curve(y_test, dummy_model_prob):

Purpose: This line calculates the False Positive Rate (FPR) and True Positive Rate (TPR) for the dummy_model. The _ is a convention in Python to indicate that the third returned value (thresholds) is not needed and will be ignored.

Mathematical Intuition: Given that dummy_model_prob consists only of zeros, this dummy model will classify all instances as the negative class (assuming a threshold like 0.5).

If there are positive instances in y_test: TP will be 0, so dummy_tpr (Recall) will be 0. FP will also be 0. So, dummy_fpr will be 0. This results in the point (0,0) on the ROC curve. The roc_curve function might return one more point depending on implementation details for edge cases, typically resulting in a curve that starts at (0,0) and ends at (0,0) or (1,0) if there are no true positives predicted.

If the dummy model always predicts class 0 and class 0 is the majority class, it will result in a curve close to the bottom-left corner of the ROC plot, or simply the point (0,0) if no positives are predicted.

model_fpr, model_tpr, thresholds = roc_curve(y_test, model_prob):

Purpose: This line calculates the FPR, TPR, and the corresponding probability thresholds for the LogisticRegression model. y_test provides the true labels, and model_prob provides the predicted probabilities for the positive class.

Mathematical Intuition: The roc_curve function works by sorting the predicted probabilities (model_prob) and then iterating through them. At each unique probability value (or a set of thresholds), it considers that value as the classification threshold. For each threshold, it calculates:

TPR (True Positive Rate): The proportion of actual positive samples that are correctly identified as positive. TPR=
fracTPTP+FN

FPR (False Positive Rate): The proportion of actual negative samples that are incorrectly identified as positive. FPR=
fracFPFP+TN
These (FPR, TPR) pairs form the points that make up the ROC curve. The thresholds array contains the specific probability values that correspond to each (FPR, TPR) pair.
