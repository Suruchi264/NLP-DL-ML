```python
## import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = 6 * np.random.rand(100,1) - 3
Y = 0.5 * X**2 + 1.5*X + 2 + np.random.rand(100,1)
## quadratic eq used = y = 0.5x^2 + 1.5x + 2 + outliers
plt.scatter(X,Y, color = 'green')
plt.xlabel('X dataset')
plt.ylabel('Y dataset')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 42)
```

```python
X_train
```

```python
## let's implement SLR
from sklearn.linear_model import LinearRegression
regression_one = LinearRegression()
```

```python
regression_one.fit(X_train, Y_train)
```

```python
from sklearn.metrics import r2_score
score = r2_score(Y_test, regression_one.predict(X_test))
print(score)
```

```python
## let's visualize this model
plt.plot(X_train, regression_one.predict(X_train), color = 'red')
plt.scatter(X_train, Y_train)
plt.xlabel('X dataset')
plt.ylabel('Y dataset')
```

```python
## let's apply polynomial transformation
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg = PolynomialFeatures(degree = 2,include_bias=True)
X_train_poly = poly_reg.fit_transform(X_train)
X_test_poly = poly_reg.transform(X_test)
```

```python
X_train_poly
```

```python
X_test_poly
```

```python
## include bias parameter
from sklearn.metrics import r2_score
regression = LinearRegression()
regression.fit(X_train_poly, Y_train)
Y_pred = regression.predict(X_test_poly)
score = r2_score(Y_test, Y_pred)
print(score)
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
plt.scatter(X_train, regression.predict(X_train_poly))
plt.scatter(X_train, Y_train)
plt.xlabel('X dataset')
plt.ylabel('Y dataset')
```

```python
poly_reg = PolynomialFeatures(degree = 3,include_bias=True)
X_train_poly = poly_reg.fit_transform(X_train)
X_test_poly = poly_reg.transform(X_test)
```

```python
X_train_poly
```

```python
## include bias parameter
from sklearn.metrics import r2_score
regression = LinearRegression()
regression.fit(X_train_poly, Y_train)
Y_pred = regression.predict(X_test_poly)
score = r2_score(Y_test, Y_pred)
print(score)
```

```python
## prediction of new dataset
X_new = np.linspace(-3,3,200).reshape(200,1)
X_new_poly = poly_reg.transform(X_new)
X_new_poly
```

```python
Y_new = regression.predict(X_new_poly)
plt.plot(X_new, Y_new, "r-", linewidth = 2, label = "New Predictions")
plt.plot(X_train, Y_train, "b.", label = "Training data")
plt.plot(X_test, Y_test, "g.", label = "Test data")
plt.legend(loc = "upper left")
plt.xlabel("X dataset")
plt.ylabel("Y dataset")
plt.legend()
plt.show()
```
