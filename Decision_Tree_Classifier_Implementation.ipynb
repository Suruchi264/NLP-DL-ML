## Decision Tree Classifier Implementation

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
print(iris['DESCR'])
```

```python
print(iris['target'])
```

```python
##independent features
X=pd.DataFrame(iris['data'],columns=['sepal length in cm','sepal width','petal length','petal width'])
```

```python
X
```

```python
##dependent features
y=iris['target']
```

```python
##train test plot
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
## apply Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
treeclassifier=DecisionTreeClassifier()
```

```python
treeclassifier.fit(X_train,y_train)
```

```python
##Visualize the Decision Tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python
y_pred=treeclassifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

## Decision Tree Classifier Prepruning And Hyperparameter Tuning

```python
param={
    'criterion':['gini','entropy', 'log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2']
}
```

Meaning:
This dictionary defines the hyperparameters and their possible values for tuning the Decision Tree.

criterion: The function to measure the quality of a split.
'gini': Gini impurity.
'entropy': Information gain (entropy).
'log_loss': Log loss (for probabilistic splits).
splitter: The strategy used to choose the split at each node.
'best': Chooses the best split.
'random': Chooses a random split.
max_depth: The maximum depth of the tree (how many splits from root to leaf).
max_features: The number of features to consider when looking for the best split.
'auto', 'sqrt', 'log2': Different strategies for feature selection.
Purpose:
To provide a grid of hyperparameters for GridSearchCV to search over and find the best combination.

Mathematical Intuition:
Hyperparameters control the complexity and behavior of the tree, affecting bias-variance tradeoff and generalization.

```python
param
```

Meaning:
Displays the parameter grid.

Purpose:
To verify the hyperparameter grid before running the search.

```python
from sklearn.model_selection import GridSearchCV
```

Meaning:
Imports GridSearchCV, a tool for exhaustive search over specified parameter values for an estimator.

Purpose:
To automate the process of finding the best hyperparameters using cross-validation.

```python
treemodel=DecisionTreeClassifier()
```

Meaning:
Initializes a Decision Tree Classifier.

Purpose:
To provide a model instance for GridSearchCV to tune.

```python
grid=GridSearchCV(treeclassifier,param_grid=param,cv=5,scoring='accuracy')
```

```python
grid.fit(X_train, y_train)
```

Meaning:

treeclassifier: The base Decision Tree model to tune.
param_grid=param: The hyperparameter grid to search.
cv=5: 5-fold cross-validation (splits data into 5 parts, trains on 4, tests on 1, repeats).
scoring='accuracy': Uses accuracy as the metric to evaluate performance.
Purpose:
To systematically try all combinations of the specified hyperparameters and find the one that gives the best cross-validated accuracy.

Mathematical Intuition:
Cross-validation helps estimate how well the model generalizes to unseen data, reducing the risk of overfitting to the training set.

```python
grid.best_params_
```

Meaning:
Returns the combination of hyperparameters that gave the best results during cross-validation.

Purpose:
To know which settings to use for the best model performance.

```python
grid.best_score_
```

Meaning:
Returns the best cross-validated accuracy score achieved during the grid search.

Purpose:
To see how well the best model performed during cross-validation.

```python
y_pred=grid.predict(X_test)
```

Meaning:
Uses the best-found model to predict the labels for the test set.

Purpose:
To evaluate the tuned model on unseen data.

```python
y_pred
```

Meaning:
Displays the predicted labels for the test set.

Purpose:
To inspect the model’s predictions.

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

Meaning:

confusion_matrix(y_test, y_pred): Shows the number of correct and incorrect predictions for each class.
classification_report(y_test, y_pred): Prints precision, recall, f1-score, and support for each class.
Purpose:
To get a detailed breakdown of the model’s performance, including types of errors.

Mathematical Intuition:

Confusion Matrix:
Rows: Actual classes
Columns: Predicted classes
Diagonal: Correct predictions
Off-diagonal: Misclassifications
Precision: ( \frac{TP}{TP+FP} )
Recall: ( \frac{TP}{TP+FN} )
F1-score: Harmonic mean of precision and recall

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_pred)
```

Meaning:

Imports the accuracy metric.
Calculates the proportion of correct predictions.
Purpose:
To get a single summary metric of model performance.

Mathematical Intuition:
[ \text{Accuracy} = \frac{\text{Number of correct predictions}}{\text{Total predictions}} ]
