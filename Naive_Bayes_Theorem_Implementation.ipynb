## Naive Baye's Machine Learning Algorithms Implementation

```python
from sklearn.datasets import load_iris
```

Meaning:
Imports the load_iris function from scikit-learn’s datasets module.

Purpose:
To load the famous Iris flower dataset, which is commonly used for classification tasks.

```python
from sklearn.model_selection import train_test_split
```

Meaning:
Imports the train_test_split function.

Purpose:
To split the dataset into training and testing sets, which is essential for evaluating model performance on unseen data.

```python
X,y=load_iris(return_X_y=True)
```

Loading the dataset

Meaning:
Loads the Iris dataset and assigns the features to X and the target labels to y.

Purpose:
To have the input features (X, a 2D array of flower measurements) and the target classes (y, a 1D array of species labels) ready for modeling.

Mathematical Intuition:

X is a matrix of shape (150, 4), where each row is a sample and each column is a feature (sepal length, sepal width, petal length, petal width).
y is a vector of length 150, with integer values representing the species.

```python
y
```

Meaning:
Displays the target labels.

Purpose:
To inspect the class labels (0, 1, 2 for the three Iris species).

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

Splitting the data


Meaning:
Splits the data into training and testing sets. 70% for training, 30% for testing.

Purpose:
To train the model on one part of the data and evaluate it on another, ensuring a fair assessment of model performance.

Mathematical Intuition:
Prevents overfitting by testing the model on data it hasn’t seen before.

```python
X_train
```

Meaning:
Displays the training features.

Purpose:
To inspect the training data.

```python
from sklearn.naive_bayes import GaussianNB
```

Meaning:
Imports the Gaussian Naive Bayes classifier.

Purpose:
To use a probabilistic classifier that assumes features are normally distributed and independent given the class.

```python
gnb=GaussianNB()
```

Meaning:
Creates an instance of the Gaussian Naive Bayes classifier.

Purpose:
Prepares the classifier for training.

```python
gnb.fit(X_train,y_train)
```

Training the model


Meaning:
Fits the Naive Bayes model to the training data.

Purpose:
The model learns the mean and variance of each feature for each class.

Mathematical Intuition:
For each class ( c ) and feature ( x_j ), it estimates: [ P(x_j | y = c) = \text{Normal}(x_j; \mu_{jc}, \sigma_{jc}^2) ] and uses Bayes’ theorem to compute the probability of each class given the features.

```python
y_pred=gnb.predict(X_test)
```

Making the predictions


Meaning:
Predicts the class labels for the test set.

Purpose:
To see how well the model generalizes to new, unseen data.

Mathematical Intuition:
For each test sample, the model computes the posterior probability for each class and picks the class with the highest probability.

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

Evaluate the model


Meaning:
Imports functions to evaluate the model’s predictions.

Purpose:
To measure the model’s performance using various metrics.

```python
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

Meaning:

confusion_matrix(y_pred, y_test): Shows the number of correct and incorrect predictions for each class.
accuracy_score(y_pred, y_test): Calculates the proportion of correct predictions.
classification_report(y_pred, y_test): Provides precision, recall, f1-score, and support for each class.
Purpose:
To get a detailed understanding of how well the model is performing.

Mathematical Intuition:

Accuracy: ( \frac{\text{Number of correct predictions}}{\text{Total predictions}} )
Precision, Recall, F1-score: Standard classification metrics that help understand the types of errors the model makes.
